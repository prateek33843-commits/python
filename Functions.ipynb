```python
# Write a function to swap two numbers without using third variable.
def swap(a,b):
    a,b=b,a
    print(a)
    print(b)
```

```python
swap(2,4)
```

```python
# Write a function to swap two numbers using third variable.
def swap(a,b):
   temp=a
   a=b
   b=temp
   print(a)
   print(b)
```

```python
swap(3,7)
```
